```python
import itertools

# Definir el grafo como una lista de adyacencia
grafo = {
    1: {2: 10, 3: 15, 4: 20, 5: 25, 6: 30, 7: 35, 8: 40},
    2: {1: 10, 3: 35, 4: 25, 5: 20, 6: 15, 7: 45, 8: 50},
    3: {1: 15, 2: 35, 4: 30, 5: 35, 6: 40, 7: 55, 8: 60},
    4: {1: 20, 2: 25, 3: 30, 5: 45, 6: 50, 7: 65, 8: 70},
    5: {1: 25, 2: 20, 3: 35, 4: 45, 6: 55, 7: 75, 8: 80},
    6: {1: 30, 2: 15, 3: 40, 4: 50, 5: 55, 7: 85, 8: 90},
    7: {1: 35, 2: 45, 3: 55, 4: 65, 5: 75, 6: 85, 8: 100},
    8: {1: 40, 2: 50, 3: 60, 4: 70, 5: 80, 6: 90, 7: 100}
}

# Obtener todos los nodos del grafo
nodos = list(grafo.keys())

# Obtener todas las permutaciones de los nodos
permutaciones = itertools.permutations(nodos)

# Imprimir todas las permutaciones como posibles caminos
for camino in permutaciones:
    print(camino)
```
